### 30. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
#first encoding gender male as 1 and female as 0
users['gender_num'] = np.where(users['gender'] == 'M' ,1,0)

#calculating number of male ratio per occupation
male_ratio = users[['occupation','gender_num']].groupby(by='occupation').mean().reset_index()
#sorting frome most ot least 
male_ratio_sorted = male_ratio[['occupation','gender_num']].sort_values(by=['gender_num'],ascending=False).reset_index()
print("Male Ratio per occupation , sorted from most to least :"'\n',male_ratio_sorted)
```

### 33.  For each occupation present the percentage of women and men

```python
# total occupatio count 
occupation_count = users['occupation'].value_counts()
#total no of men 
men_count = users[users['gender']=='M']['occupation'].value_counts()
# total no of women 
women_count = users[users['gender'] == 'F']['occupation'].value_counts()
#creating a dataframe for calculating the percentage of men and women per occupation 
occupation_gender_ratio = pd.DataFrame({
    'Total': occupation_count,
    'Men':men_count,
    "Women":women_count
})


occupation_gender_ratio = occupation_gender_ratio.fillna(0)# for null value assinging 0
# Caculating persentage of men and women 
occupation_gender_ratio['Men %']= occupation_gender_ratio['Men']/occupation_gender_ratio['Total']*100

occupation_gender_ratio['Women %']= occupation_gender_ratio['Women']/occupation_gender_ratio['Total']*100

# displaing 

occupation_gender_ratio[['Men %','Women %']].reset_index()
```

### 49. Write a Python script to display the
- a. Current date and time
- b. Current year
- c. Month of year
- d. Week number of the year
- e. Weekday of the week
- f. Day of year
- g. Day of the month
- h. Day of week

```python
from datetime import datetime, timedelta

# Part a: Current date and time
current_datetime = datetime.now()
print("Current date and time:", current_datetime)

# Part b: Current year
current_year = current_datetime.year
print("Current year:", current_year)

# Part c: Month of year
current_month = current_datetime.strftime('%B')
print("Month of year:", current_month)

# Part d: Week number of the year
week_number = current_datetime.strftime('%U')
print("Week number of the year:", week_number)

# Part e: Weekday of the week
weekday = current_datetime.strftime('%A')
print("Weekday of the week:", weekday)

# Part f: Day of year
day_of_year = current_datetime.strftime('%j')
print("Day of year:", day_of_year)

# Part g: Day of the month
day_of_month = current_datetime.strftime('%d')
print("Day of the month:", day_of_month)

# Part h: Day of week
day_of_week = current_datetime.strftime('%w')
print("Day of week:", day_of_week)
```
